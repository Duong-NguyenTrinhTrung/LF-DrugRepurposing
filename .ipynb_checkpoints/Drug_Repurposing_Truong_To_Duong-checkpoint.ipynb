```python
df12000test1 = pd.read_csv('12000test1.csv')
```

```python
type(df12000test1)
```

```python
df12000test1.Bioactivity.replace(to_replace=dict(active=1, inactive=0), inplace=True)
```

```python
df12000test1
```

```python
X = df12000test1.drop(['Bioactivity'], axis=1)
Y = df12000test1['Bioactivity']
```

```python
# Split the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = \
train_test_split(X, Y, test_size = 0.2, random_state = 1234, stratify=Y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Define the keras sequential model with three hidden layers
model = Sequential()
model.add(Dense(24,
                input_shape=(881,),
                activation='relu',
                kernel_initializer='RandomNormal'))

model.add(Dense(12,
                activation='relu',
                kernel_initializer='RandomNormal'))

model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the keras model for classification accuracy
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Fit the model on the training dataset
model.fit( X_train, Y_train, epochs=16, batch_size=10,verbose=0, )
```

```python
# Evaluate and print the model accuracy on test dataset
accuracy_test = model.evaluate(X_test, Y_test)
print("")
print("Test Accuracy     : " + str(accuracy_test[1]))
```

```python
Y_predict = (model.predict(X_test) > 0.5).astype('int32')
Y_pred_prob = model.predict(X_test)
```

```python
# Create the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y_test, Y_predict)
```
